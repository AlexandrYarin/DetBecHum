```python
import pyautogui as pa
import time
import random 
```

```python
REGION = (1550, 50, 40, 700)
REGION_ = (1555, 463, 35, 30)
OFFSET = 50
DELAY = (0.03, 0.07)
CONFIDENCE = 0.999
VAL_GAMES = 7
GAME_BUTTON = (1695, 771)
```

```python
tab = pa.screenshot(region=REGION_)
tab.save('pages\\tabizoo.png')
```

```python
pa.position()
```

```python
pa.click(1856, 200)
```

```python
x,y = pa.locateCenterOnScreen('pages\\tabizoo.png', region=REGION)
pa.moveTo(x,y, duration=0.5)
```

```python
pa.screenshot(region=REGION)
```

```python
for i in ['poket.png', 'blum.png', 'yes.png']:
    x,y = pa.locateCenterOnScreen(i, region=REGION, confidence=CONFIDENCE)
    pa.moveTo(x,y)
    time.sleep(2)
```

```python
for i in range(VAL_GAMES):
    pa.click(*GAME_BUTTON)
    time.sleep(2)
    end_time = time.time() + 28

    while time.time() <= end_time:
        try:
            x, y = pa.locateCenterOnScreen('star4.png', region=REGION, confidence=CONFIDENCE, grayscale=True)
            pa.click(x, y + OFFSET)
            time.sleep(random.uniform(*DELAY))
        except pa.ImageNotFoundException:
            print('image not found')
            continue
    time.sleep(random.randrange(3,6))  
```
